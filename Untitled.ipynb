```python
def factorial(n):
  if n==0:
    return 1
  else:
    return n * factorial(n-1)

num = int(input("Enter a number: "))
print("Factorial of", num, "is", factorial(num))
```

```python
def evenodd(a):
  if a%2==0:
    return('even')
  else:
    return('odd')

result =  int(input("Enter a number: "))
print ("entered number", result, "is", evenodd(result))
```

```python
def calculate_mean(numbers):
    return sum(numbers)/len(numbers)

def calculate_median(numbers):
    sorted_numbers=sorted(numbers)
    n=len(sorted_numbers)
    middle=n//2

    if n%2==0:
        return (sorted_numbers[middle-1]+sorted_numbers[middle]/2)
    else:
        return (sorted_numbers[middle])

numbers=[1, 3, 3, 6, 7, 8, 9]
mean=calculate_mean(numbers)
median=calculate_median(numbers)
print("mean",mean)
print("median",median)
```

```python
import matplotlib.pyplot as plt
import numpy as np

hours_studied = [10, 9, 2, 15, 10, 16, 11, 16]
exam_scores = [95, 80, 10, 50, 45, 98, 38, 93]
x=np.array(hours_studied)
y=np.array(exam_scores)
x,y=zip(*sorted(zip(x,y)))
plt.plot(x,y, marker="*", color="blue", linestyle="--")
plt.xlabel("hours studied")
plt.ylabel("exam score")
plt.title( "hours vs exam")
plt.grid(True)
plt.show()
```
